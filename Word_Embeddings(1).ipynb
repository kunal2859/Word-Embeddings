**Bag of Words**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"text":["people watch dswithkunal",
                         "dswithkunal watch dswithkunal",
                         "people write comment",
                          "dswithkunal write comment"],"output":[1,1,0,0]})

df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
bow=cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
print(bow[3].toarray())
```

```python
cv.transform(['Kunal watch dswithkunal']).toarray()
```

```python
x=bow.toarray()
y=df['output']
```

**N-Grams**

```python
df
```

```python
#bi grams
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(2,2))
```

```python
ngram=cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(ngram[0].toarray())
print(ngram[1].toarray())
print(ngram[2].toarray())
print(ngram[3].toarray())
```
